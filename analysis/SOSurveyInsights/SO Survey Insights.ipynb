## <a id='Philosophy'>Philosophy</a>
The philosophy of this analysis is to uncover certain not-so-obvious insights reg. Inclusiveness, Community Companionship and Gender of Stack Overflow Members and Developers in General.
#### Focus Areas:

* Salary and Salary Gap


* Companionship


* Gender-focused Dimensions Analysis

## <a id='AboutTheDataset'>About the dataset</a>
This is the response data of Stack Overflow 2018 Developer Survey.

Number of Qualified Respondents: 98,855 Number of Completed Respondents: 67,441

More details <a href='https://www.kaggle.com/stackoverflow/stack-overflow-2018-developer-survey'>here</a>

## <a id='Highlights'>Highlights</a>
* Stack Overflow (overall) boasts an **NPS of 86** that’s off the chart in the world of Net Promoter Scores - to make some sense, Amazon’s NPS according to npsbenchmarks.com is 69.

* **Female Developers** are insanely less with just **7%** of the entire respondents being female against 92% Male Developers.

* **58% of Computer Science/Engineering and Health Science** Undergraduate Majors consider themselves to be part of Stack Overflow but only **44% of Fine Arts and 45% of Humanities & Social Science** Undergrad Majors consider to be part of Stack Overflow.

* **Mobile Developers** and **Students** are on the opposite spectrum of Feeling inclusive in Stack Overflow with only less than 50% of students feeling inclusive.

* Developers of **Python, Scala, Perl, Bash/Shell** are in the middle of this spectrum of Feeling Inclusive and Participating Frequently.

* On an index of No_to_Yes Ratio for feeling Inclusive, **Finland, Norway, New Zealand, Canada and UK** lead the chart.

* **~19K developers** either disagree or neigher disagree nor agree, which is a significant number that there’s still a decent room to improve the community companionship.

* **35% of developers** Agree thinking that they are competing while **38% disagree**.

* Overall, The effect of **imposter syndrome** seems to be minimal among the respondents with **38K respondents** either disagreeing or strongly disagreeing it.

* Imposter Syndrome (Feeling of not good at programming) looks like that this effect is more prevelant among Female Develpers than their male counterparts comparatively.

* Female Developers in **Product manager, Full-stack developer, System administrator** domains are getting almost equal salary compared to Male.

* The first mover advantage for Male Respondents could be seen here. **~20% of Male in the age bracket of 18-24** are **independent contractors / freelancer/self-employed** while It’s **~16% of Female** in that range.

* Among Indie Contractors/Freelancers, It is **Female with Bachelor’s 43% and Master’s 26% lead** their group than their equivalent **Male counterpart whose Bachelor’s is 38% and Master’s is 23%**. This could be possibly because of the fact that college-educated female developers tend to take a **freelancer role or a contractual role due to the flexibility** it provides.

### <a id='SurveyCountries'>Countries</a>
#### <a id='MostRepresentedCountries'>Most Represented Countries</a>

```python
country_not_null = survey18[survey18['Country'].isnull() == False]
most_represented_countries = pd.DataFrame(country_not_null.groupby(['Country']).size().reset_index(name='RepresentationCount'))
most_represented_countries = most_represented_countries.sort_values(by = ['RepresentationCount'], ascending = False)
most_represented_countries.head(10)
```

### <a id='FemaleToMaleRatio'>Female to Male Ratio</a>
#### <a id='FemaleToMaleRatioBetterCountries'>Female-to-Male Ratio - Better Countries</a>
Imagine, we live in a world where for Countries with more than 200 developers (from the respondents), 0.16 is the best Female to Male Ratio.

```python
gender_and_country_df = pd.merge(most_represented_countries, survey18[['Country','Gender']], on='Country')
gender_and_country_df = gender_and_country_df[gender_and_country_df.RepresentationCount>200]
mratio = pd.DataFrame(gender_and_country_df.groupby([gender_and_country_df.Country,gender_and_country_df.Gender[gender_and_country_df.Gender=='Male']]).size().reset_index(name='MaleCount'))
fratio = pd.DataFrame(gender_and_country_df.groupby([gender_and_country_df.Country,gender_and_country_df.Gender[gender_and_country_df.Gender=='Female']]).size().reset_index(name='FemaleCount'))
f2mratio = pd.merge(mratio.drop('Gender', axis=1), fratio.drop('Gender', axis=1))
f2mratio['F2MRatio'] = f2mratio['FemaleCount']/f2mratio['MaleCount']
f2mratio = f2mratio.sort_values(by = 'F2MRatio', ascending = False)
f2mratio.head(10)
```

### <a id='Gender'>Gender</a>
With no doubt, the respondents are primarily Male with 92% and Female being just 7%

```python
gender = survey18['Gender'].copy()
gender = gender.dropna()
gender_series_without_semicolon = gender[~(gender.str.contains(";"))].value_counts()
gender_series_with_semicolon = gender[gender.str.contains(";")].str.split(';', expand=True).stack().value_counts()
gender_count = (gender_series_without_semicolon + gender_series_with_semicolon).sort_values(ascending=False)
gender_percentage = gender_count.to_frame(name='Count')
gender_percentage['Percentage'] =  round((gender_percentage.Count/sum(gender_percentage.Count))*100).astype(int)
gender_percentage
```

### <a id='AnnualSalary'>Annual Salary (USD)</a>
#### <a id='OverallDistribution'>Overall Distribution</a>

```python
salary_category_list = []
salary = survey18['ConvertedSalary'].copy()
salary = salary.replace('0',np.nan)
salary = salary.dropna()
salary_category_list.append(salary[salary.astype(float)<=500000].shape[0])
salary_category_list.append(salary[(salary.astype(float)>500000) & (salary.astype(float)<=1000000)].shape[0])
salary_category_list.append(salary[(salary.astype(float)>1000000) & (salary.astype(float)<=1500000)].shape[0])
salary_category_list.append(salary[(salary.astype(float)>1500000) & (salary.astype(float)<=2000000)].shape[0])
count_by_category = pd.Series(salary_category_list)
count_by_category.set_axis(['Salary<500000', '500000-1000000', '1000000-1500000','1500000-2000000'], axis=0, inplace=True)
count_by_category
```

```python
net_promoter_score = survey18['StackOverflowJobsRecommend'].copy().dropna()
net_promoter_score = net_promoter_score.replace('10 (Very Likely)','10')
net_promoter_score = net_promoter_score.replace('0 (Not Likely)','0')
net_promoter_score = net_promoter_score.astype(int)
net_promoter_score = net_promoter_score.value_counts()
net_promoter_score = net_promoter_score.sort_index()
net_promoter_score
```

```python
passive_sum, detractor_sum, promoter_sum = 0, 0, 0
list_of_percentages = []

for i, v in net_promoter_score.iteritems():
    if i < 7:
        detractor_sum = detractor_sum + v
    elif i >= 9:
        promoter_sum = promoter_sum + v
    else:
        passive_sum = passive_sum + v
    
total_sum = detractor_sum + promoter_sum + passive_sum
detractor_percentage = (detractor_sum/total_sum) * 100
promoter_percentage = (promoter_sum/total_sum) * 100
passive_percentage = (passive_sum/total_sum) * 100

list_of_percentages = [promoter_percentage, passive_percentage, detractor_percentage]
percentage_by_type = pd.Series(list_of_percentages)
percentage_by_type = percentage_by_type.set_axis(['promoter_percentage', 'passive_percentage', 'detractor_percentage'],inplace=False)
percentage_by_type
```

## <a id='NetPromoterScore'>Net Promoter Score</a>


### <a id='WhatIsNPS'>What is NPS?</a>
The Net Promoter Score is an index ranging from -100 to 100 that measures the willingness of customers to recommend a company’s products or services to others. It is used as a proxy for gauging the customer’s overall satisfaction with a company’s product or service and the customer’s loyalty to the brand.

Source: https://www.medallia.com/net-promoter-score/

```python
net_promoter_score = survey18['StackOverflowRecommend'].copy().dropna()
net_promoter_score = net_promoter_score.replace('10 (Very Likely)','10')
net_promoter_score = net_promoter_score.replace('0 (Not Likely)','0')
net_promoter_score = net_promoter_score.astype(int)
net_promoter_score = net_promoter_score.value_counts()
net_promoter_score = net_promoter_score.sort_index()
net_promoter_score
```

### <a id='StackOverflowNPS'>Stack Overflow NPS</a>


```python
passive_sum, detractor_sum, promoter_sum = 0, 0, 0
list_of_percentages = []

for i, v in net_promoter_score.iteritems():
    if i < 7:
        detractor_sum = detractor_sum + v
    elif i >= 9:
        promoter_sum = promoter_sum + v
    else:
        passive_sum = passive_sum + v
    
total_sum = detractor_sum + promoter_sum + passive_sum
detractor_percentage = (detractor_sum/total_sum) * 100
promoter_percentage = (promoter_sum/total_sum) * 100
passive_percentage = (passive_sum/total_sum) * 100

list_of_percentages = [promoter_percentage, passive_percentage, detractor_percentage]
percentage_by_type = pd.Series(list_of_percentages)
percentage_by_type = percentage_by_type.set_axis(['promoter_percentage', 'passive_percentage', 'detractor_percentage'],inplace=False)
percentage_by_type
```

**Net Promoter Score for Stack Overflow Jobs is:**

```python
promoter_percentage - detractor_percentage
```

### <a id='VisitFrequencyAndnclusiveness'>NPS ~ Visit Frequency + Inclusiveness</a>

It can been seen from the below plot that even most of those who don’t feel prat of the community still would like to recommend Stack Oveflow to their friends/family.

```python
visit_frequency = survey18[['StackOverflowRecommend','StackOverflowVisit','StackOverflowConsiderMember']].copy().dropna()
visit_frequency = visit_frequency[visit_frequency.StackOverflowVisit != 'I have never visited Stack Overflow (before today)']
visit_frequency['StackOverflowRecommend'] = visit_frequency.StackOverflowRecommend.replace('10 (Very Likely)','10')
visit_frequency['StackOverflowRecommend'] = visit_frequency.StackOverflowRecommend.replace('0 (Not Likely)','0')

def label(row):
    if int(row['StackOverflowRecommend']) >=9:
        val = 'Promoter'
    elif int(row['StackOverflowRecommend']) < 7:
        val = 'Detractor'
    else:
        val = 'Passive'
    return val

visit_frequency['StackOverflowRecommendLabel'] = visit_frequency.apply(label, axis=1)
visit_frequency = visit_frequency.groupby(['StackOverflowVisit','StackOverflowConsiderMember','StackOverflowRecommendLabel']).size().reset_index(name='Count')
visit_frequency = visit_frequency.sort_values(by = 'Count', ascending = False)
visit_frequency.head(10)
```
