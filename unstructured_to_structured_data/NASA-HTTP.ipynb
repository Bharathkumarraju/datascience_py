```python
structureddf = pd.DataFrame
total_rows = 0
counter = 1
for unstructureddf in pd.read_csv('datasets/NASA_access_log_Aug95', header=None, delimiter='\n', encoding = "iso-8859-1", iterator=True, chunksize=50000):
    structureddf = unstructureddf[0].str.extract("^(.*?)\s-\s-",expand=True)
    structureddf[1] = unstructureddf[0].str.extract("\[(.*)\]",expand=True)
    structureddf[2] = unstructureddf[0].str.extract('\"(.*)\"',expand=True)
    structureddf[3] = unstructureddf[0].str.extract('\s(200|302|304|400|404)\s',expand=True)
    structureddf[4] = unstructureddf[0].str.extract('(\d+)(?!.*\d)',expand=True)
    counter+=1
    total_rows+=structureddf.shape[0]
total_rows
```

```python
structureddf.rename(index=str, columns={0: "hostname", 1: "timestamp", 2: "http_request", 3: "http_reply_code", 4: "bytes_in_reply"}) 
```
