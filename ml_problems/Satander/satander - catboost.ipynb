```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
y_train = train.pop('target').values
train['source']='train'
test['source']='test'
df = pd.concat([train, test],ignore_index=True)
df = df.set_index('ID_code')
df.head()
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
X_train = df[df['source_train']==1].drop(['source_train','source_test'], axis=1).values
X_test = df[df['source_test']==1].drop(['source_train','source_test'], axis=1).values
```

```python
import catboost as cb

model = cb.CatBoostClassifier(eval_metric="AUC", depth=10, iterations= 500, l2_leaf_reg= 9, learning_rate= 0.15)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
sample_file = pd.read_csv('sample_submission.csv')
sample_file['target'] = y_pred
sample_file.to_csv('satander_submission3.csv', index=False)
```
